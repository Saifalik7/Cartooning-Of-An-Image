```python
from imutils import paths
import numpy as np
import cv2
import os
```

```python
imagepaths = list(paths.list_images("Main"))
```

```python

instances = []

# Load in the images
for filepath in os.listdir('Main'):
    instances.append(cv2.imread('Main/{0}'.format(filepath),0))

#print(type(instances[0]))
from matplotlib import pyplot as plt
plt.imshow(instances[399])
```

```python

x = len(imagepaths)
x
```

```python
images = []

for i in instances:
    images.append(cv2.cvtColor(i, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(images[399])
```

```python
img_small = []

for i in images:
    img_small.append(cv2.pyrDown(i))
    
plt.imshow(img_small[0])
```

```python
imagesBil = []

for i in img_small:
    num_iter = 5
    for _ in range(num_iter):
         img = cv2.bilateralFilter(i, d=9, sigmaColor=9, sigmaSpace=7)
    imagesBil.append(img)
            
plt.imshow(imagesBil[0])
```

```python
img_rgb = []

for i in imagesBil:
    img_rgb.append(cv2.pyrUp(i))
    
plt.imshow(img_rgb[399])
```

# Edge Detection

```python
img_gray = []

for i in img_rgb:
    img_gray.append(cv2.cvtColor(i, cv2.COLOR_RGB2GRAY))
```

```python
img_blur = []

for i in img_gray:
    img_blur.append(cv2.medianBlur(i, 7))
    
plt.imshow(img_blur[399])
```

```python
img_edge = []

for i in img_blur:
    img_edge.append(cv2.adaptiveThreshold(i, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7, 2))
  
plt.imshow(img_edge[399])

    
```

```python
img_togray = []

for i in img_edge:
    img_togray.append(cv2.cvtColor(i, cv2.COLOR_GRAY2RGB))
    
plt.imshow(img_togray[399])
```

```python
result = []

for i,j in zip(img_rgb, img_togray):
    result.append( cv2.bitwise_and(i, j))
```

```python
plt.imshow(result[399])
```
